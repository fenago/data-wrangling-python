0. load dataset

```python
import numpy as np 

import pandas as pd 

import matplotlib.pyplot as plt 

df = pd.read_excel("Sample - Superstore.xls") 

df.head() 
```

1. Sample 4 records each to create three DataFrames at random from the original sales dataset

```python
df_1 = df[['Customer Name','State','Sales','Profit']].sample(n=4) 

df_2 = df[['Customer Name','State','Sales','Profit']].sample(n=4) 

df_3 = df[['Customer Name','State','Sales','Profit']].sample(n=4) 
```

2. combine all three dataframes

```python
df_cat2 = pd.concat([df_1,df_2,df_3], axis=1) 

df_cat2 
```

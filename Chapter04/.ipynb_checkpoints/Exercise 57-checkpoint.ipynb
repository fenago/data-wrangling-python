```python
import numpy as np 

import pandas as pd 

import matplotlib.pyplot as plt 

df = pd.read_excel("Sample - Superstore.xls") 

df.head() 
```

1.Specify the number of samples that you require from the DataFrame

```python
df.sample(n=5) 
```

```python
df.sample(frac=0.1) 
```

```python
df.sample(frac=0.1, replace=True) 
```
